```python
#7 Build a book recommendation engine to recommend books

# About Book Crossing Dataset

# Dataset comprises of three tables for users, books and ratings. Explicit ratings are expressed on a scale from 1-10 (higher values denoting higher appreciation) and implicit rating is expressed by 0
#Link to dataset files
#http://www2.informatik.uni-freiburg.de/~cziegler/BX/
# Download following files - 
# The Book-Crossing dataset comprises 3 tables.

#     BX-Users
#     Contains the users. Note that user IDs (`User-ID`) have been anonymized and map to integers. Demographic data is provided (`Location`, `Age`) if available. Otherwise, these fields contain NULL-values.

#     BX-Books
#     Books are identified by their respective ISBN. Invalid ISBNs have already been removed from the dataset. Moreover, some content-based information is given (`Book-Title`, `Book-Author`, `Year-Of-Publication`, `Publisher`), obtained from Amazon Web Services. Note that in case of several authors, only the first is provided. URLs linking to cover images are also given, appearing in three different flavours (`Image-URL-S`, `Image-URL-M`, `Image-URL-L`), i.e., small, medium, large. These URLs point to the Amazon web site.

#     BX-Book-Ratings
#     Contains the book rating information. Ratings (`Book-Rating`) are either explicit, expressed on a scale from 1-10 (higher values denoting higher appreciation), or implicit, expressed by 0.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
import warnings
warnings.filterwarnings("ignore")
```

```python
books = pd.read_csv('BX-Books.csv', sep=';', error_bad_lines=False, encoding="latin-1")
books.columns = ['ISBN', 'bookTitle', 'bookAuthor', 'yearOfPublication', 'publisher', 'imageUrlS', 'imageUrlM', 'imageUrlL']
users = pd.read_csv('BX-Users.csv', sep=';', error_bad_lines=False, encoding="latin-1")
users.columns = ['userID', 'Location', 'Age']
ratings = pd.read_csv('BX-Book-Ratings.csv', sep=';', error_bad_lines=False, encoding="latin-1")
ratings.columns = ['userID', 'ISBN', 'bookRating']
```

```python
book_ratings = ratings.merge(books,on='ISBN')
#book_ratings.head(2)

ratings_cnt = book_ratings.groupby('bookTitle')['bookRating'].count().reset_index().rename(columns={'bookRating':'ratings_cnt'})
avg_rating_cnt = book_ratings.groupby('bookTitle')['bookRating'].mean().reset_index().rename(columns={'bookRating':'avg_rating'})
```

### Popularity based filtering of books
#### Charting out books on 'number of ratings count' and 'maximum average ratings'

```python
popular_df = ratings_cnt.merge(avg_rating_cnt,on='bookTitle')
#popular_df

popular_df = popular_df[popular_df['ratings_cnt']>=250]

popular_df = popular_df.sort_values('avg_rating',ascending=False).head(50)
popular_df.head(10)
```

### Item based collaborative filtering

#### Checking for the statistically significant threshold for users with count of ratings, and books with count of ratings

```python
bkrt=book_ratings.groupby('userID')['bookRating'].count().reset_index().rename(columns={'bookRating':'cnts'})

print(bkrt.describe())

print("Percentile value:",bkrt.cnts.quantile(0.99))
```

```python
users.drop('Age',axis=1,inplace=True)
books.dropna(inplace=True)

book_ratings.dropna(inplace=True)

### Applying the threshold as >200 for count of users giving ratings
x = book_ratings.groupby('userID')['bookRating'].count()>200
experience_users = x[x].index

filtered_rating_users = book_ratings[book_ratings['userID'].isin(experience_users)]
```

```python
usrrat=filtered_rating_users.groupby('bookTitle')['bookRating'].count().reset_index().rename(columns={'bookRating':'cnts'})

print(usrrat.describe())

print("Percentile value:",usrrat.cnts.quantile(0.99))
```

```python
### Applying the threshold as >50 for count of books given a rating
y = filtered_rating_users.groupby('bookTitle')['bookRating'].count()>50
famous_books = y[y].index
```

```python
#### Books with very less counts and users with very less numbers of rating counts won't be significant for our model

## Charting out the list of famous books by applying the threshold as >50
final_ratings = filtered_rating_users[filtered_rating_users['bookTitle'].isin(famous_books)]

## Creating the final modelling dataframe
model_df = final_ratings.pivot_table(index='bookTitle',columns='userID',values='bookRating')
model_df.fillna(0,inplace=True)
model_df
```

### Cosine similarity

### Similarity = (A.B) / (||A||.||B||) 
#### For example: Calc. the cosine similarity for '1984' & '1st to Die: A Novel' - lets assume that there is only one rating as common. 
#### Formula = ((rating of '1984') x (rating of '1st to Die: A Novel') given by user) / (SQRT(Individual squares of all ratings) x SQRT(Individual squares of all ratings)) 
#### For more than one rating as common: 
#### Formula = ((rating of '1984') x (rating of '1st to Die: A Novel') + (rating of '1984') x (rating of '1st to Die: A Novel') given by individual user) / (SQRT(Individual squares of all ratings) x SQRT(Individual squares of all ratings)) of the individual users



```python
cos_scores = cosine_similarity(model_df)

def recommend(bookname):
    indexbook = np.where(model_df.index == bookname)[0][0]
    dis = cos_scores[indexbook]
    book_l = sorted(list(enumerate(dis)),reverse=True,key=lambda x:x[1])[1:7]
    
    for i in book_l:
        print(model_df.index[i[0]])       
```

### Nearest neighbour model

```python
knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
knn.fit(model_df)
```

```python
## Randomly choosing a book for the recommendation engine
query_index = np.random.choice(model_df.shape[0])
print(query_index)
```

##### Getting the cos(theta) value and then 1 - cos(theta) to get the distance

```python
## Finding 6 similar books/values using the 'n_neighbors' parameters
dist, indx = knn.kneighbors(model_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 7)
```

#### Generating the recommendations/Outputs

```python
## Using Cosine similarity
print("Recommendation for:",model_df.index[query_index])
recommend(model_df.index[query_index])
```

```python
## Using KNN 
for i in range(0, len(dist.flatten())):
    if i == 0:
        print('Recommendations for:',model_df.index[query_index])
    else:
        print('{}'.format(model_df.index[indx.flatten()[i]]))
```
